```python
task = "2"
ann_root = Path("/home.stud/kuntluka/dataset/task_dental_rtg" + task + "/annotations")
img_root = Path("/home.stud/kuntluka/dataset/task_dental_rtg" + task + "/images")
ann_name = "instances_default.json"

with open(ann_root/ann_name) as f:
    data = json.load(f)
# blacklist = list(range(962, 1000))
# blacklist = blacklist + [97, 98, 99]

# blacklist = [str(x) + ".png" for x in blacklist]
blacklist = duplicit_img_names

new_data = {}
new_data['categories'] = data['categories'][0]
new_data['images'] = []
new_data['annotations'] = []
img_ids = []
for img in data['images']:
    if img['file_name'] in blacklist:
        continue
    else:
        img_ids.append(img['id'])
        new_data['images'].append(img)

carries_cat = new_data['categories']['id']

for annotation in data['annotations']:
    if annotation['category_id'] != carries_cat:
        continue
    if annotation['image_id'] not in img_ids:
        continue
    new_data['annotations'].append(annotation)
print(f"images: old {len(data['images'])} new {len(new_data['images'])}")
print(f"annotations: old {len(data['annotations'])} new {len(new_data['annotations'])}")

with open(ann_root / 'instances.json', 'w') as f:
    json.dump(new_data, f)

# print(new_data)
```
